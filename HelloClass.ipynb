```python
###In this class, you will generate a random normal sample of 1000 members, 
###and you will compare it with theoretical distribution
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(123)
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```python
### If you are doing it correctly, you should generate a plot
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
